SUPER STORE DATA ANALYSIS
            

```python
#importing all necessary libraries

import pandas as pd
import numpy  as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#storing the dataset into "df" DATAFRAME
df=pd.read_csv("SampleSuperstore.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#checking the numbers of unique entries in each column
df.nunique()
```

```python
print(df['Country'].unique())
print(df['Ship Mode'].unique())
print(df['Region'].unique())
print(df['Category'].unique())
print(df['Sub-Category'].unique())
print(df['State'].unique())
print(df['Segment'].unique())
```

```python
gbsegment_sum=df.groupby('Segment').sum()
```

```python
gbsegment_mean=df.groupby('Segment').mean()
```

# **Below Graph shows The Total numbers of profit and sales per"REGION"**

```python
df.groupby("Region")[["Profit","Sales"]].sum().plot.bar()
plt.title(" Total Profit and Sales Per Region")
plt.rcParams['figure.figsize'] = [12,10]
plt.show()
```

from this above graph we can conclude that West region is the highest in product sales and also in profit.Although south region is the lowest in sales but central region is lowest in the profit.

# ***Below Graph shows The Total numbers of profit and sales IN "COUNTRY"***

```python
df.groupby("Country")[["Profit","Sales"]].sum().plot.bar()
plt.title(" Total Profit and Sales ")
plt.rcParams['figure.figsize'] = [12,10]
plt.show()
```

### from above the graph we can see that the hole country's total sale is high but comparatively the total profit is much lower than sales.

# Below Graph shows The Total numbers of profit and sales per"SHIP MODE"

```python
df.groupby("Ship Mode")[["Profit","Sales"]].sum().plot.area()
plt.title(" Total Profit and Sales Per Ship Mode")
plt.rcParams['figure.figsize'] = [10,8]
plt.show()
```

from the above graph we can assume that most of the sales mode is standard class and in same day ship mode lowest in profit and sales also.

# ***Below Graph shows The Total numbers of profit and sales per"STATES"***

```python
df.groupby("State")[["Profit","Sales"]].sum().plot.bar()
plt.title(" Total Profit and Sales Per State")
plt.rcParams['figure.figsize'] = [12,10]
plt.show()
```

From above graph we can understand that supermarkets in the 10 states are in loss; where as New York and California are showing huge amount of "Total Profits" where as some states are showing profits in average range while a lot of states are making a very little profits. But when we see the profits per transaction New York and California boils down to the average ranges with others while the state like Vermont it is making higher profits per transaction.

From this we can deduce that this type of behaviour in New York and California is caused due to making higher number of sales, where in Vermont the number of sales are made less but the profit made in each transaction is higher.

# **Below Graphs shows Total Number of Profit and Sales made "BY SEGMENT"**

```python
plt.figure(figsize=(14,4))
sns.set_style('whitegrid')

plt.subplot(1,2,1)
plt.title('Total Profit')
sns.barplot(x=gbsegment_sum.index,y='Profit',data=gbsegment_sum,ci=None)

plt.subplot(1,2,2)
plt.title('Total Sales')
sns.barplot(x=gbsegment_mean.index,y='Sales',data=gbsegment_mean,ci=None)
```

from the above graphs we can analyse that in segment section,home office part gains highest sales but gives lowest profit and the consumer part shows the highest profits.

# **Below Graphs shows  Profit and Sales made Per Transaction "BY SEGMENT"**

```python
plt.figure(figsize=(14,5))
sns.set_style('whitegrid')

plt.subplot(1,2,1)
plt.title('Profit per Transaction')
sns.barplot(x=gbsegment_mean.index,y='Profit',data=gbsegment_mean,ci=None)

plt.subplot(1,2,2)
plt.title('Sales Per Transaction')
sns.barplot(x=gbsegment_mean.index,y='Sales',data=gbsegment_mean,ci=None)
```
